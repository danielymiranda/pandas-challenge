```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# Create reference to CSV file
PurchaseData_csv = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(PurchaseData_csv)
purchase_data_df.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items=purchase_data_df["Item ID"].nunique()

average_price= purchase_data_df["Price"].mean()
Format_price=("${:,.4}".format(average_price))

number_purchases=purchase_data_df["Purchase ID"].count()

total_revenue=purchase_data_df["Price"].sum()
Format_revenue=("${:,}".format(total_revenue))

raw_analysis = {
    "Number of Unique Items": [unique_items],
    "Average Price": [Format_price],
    "Number of Purchases": [number_purchases],
    "Total Revenue":[Format_revenue]
}
analysis_pd = pd.DataFrame(raw_analysis, columns=["Number of Unique Items",
                                                  "Average Price", "Number of Purchases",
                                                  "Total Revenue"])
analysis_pd
```

## Age Demographics

```python
bins=[0,9,14,19,24,29,34,39,50]
group_ages=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data_df["Age Range"]=pd.cut(purchase_data_df["Age"], bins, labels=group_ages) 

groupby_range_age=purchase_data_df.groupby("Age Range")

age_range=groupby_range_age["SN"].nunique()
sum_age=pd.DataFrame({"TotalCount":age_range})

age_percent=sum_age.TotalCount/sum_age.TotalCount.sum()

#Format_age_percent=("{:.2%}".format(age_percent))

sum_age=pd.DataFrame({"TotalCount":age_range,
                     "Percentage of Players":age_percent})

formatter = {'Percentage of Players':'{:.2%}'}

sum_age

sum_age.style.format(formatter)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
Purchase_Count=groupby_range_age["Purchase ID"].count()
Average_Purchase_Price=groupby_range_age["Price"].mean()
Total_Purchase_Value=groupby_range_age["Price"].sum()
Total_PPP=groupby_range_age["SN"].nunique()

Purchasing_Analysis=pd.DataFrame({"Purchase Count":Purchase_Count,
                                "Average Purchase Price":Average_Purchase_Price,
                                 "Total Purchase Value":Total_Purchase_Value,
                                  "Average Total Purchase per Person":Total_Purchase_Value/Total_PPP})

formatter_2 = {'Average Purchase Price':'${:,.3}','Total Purchase Value':'${:,.5}',
              'Average Total Purchase per Person':'${:,.3}'}

Purchasing_Analysis

Purchasing_Analysis.style.format(formatter_2)
```

## Top Spenders

```python
groupby_range_sn=purchase_data_df.groupby("SN")

Purchase_Count_SN=groupby_range_sn["Purchase ID"].count()
Purchase_Avg_SN=groupby_range_sn["Price"].mean()
Purchase_Total_SN=groupby_range_sn["Price"].sum()

Top_Spenders_Analysis=pd.DataFrame({"Purchase Count":Purchase_Count_SN,
                                "Average Purchase Price":Purchase_Avg_SN,
                                 "Total Purchase Value":Purchase_Total_SN})

#Top_Spenders.head(5)

Top_Spenders=Top_Spenders_Analysis.sort_values("Total Purchase Value",ascending=False)

#formatter_3 = {'Average Purchase Price':'${:,.3}','Total Purchase Value':'${:,.5}'}

Top_Spenders.head(5)

#Top_Spenders.style.format(formatter_3)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
groupby_range_ID=purchase_data_df.groupby("Item ID")

Item_Name=(groupby_range_ID["Item Name"].unique()).str[0]
Purchase_Count_ID=groupby_range_ID["Purchase ID"].count()
Item_Price_ID=(groupby_range_ID["Price"].unique()).str[0]
Total_Purchase_Value_ID=groupby_range_ID["Price"].sum()

Most_Popular_Analysis=pd.DataFrame({"Item Name":Item_Name,
                                "Purchase Count":Purchase_Count_ID,
                                 "Item Price":Item_Price_ID,
                              "Total Purchase Value":Total_Purchase_Value_ID})

Most_Popular_Analysis=Most_Popular_Analysis.sort_values("Purchase Count", ascending=False)

Most_Popular_Analysis.head()
```
